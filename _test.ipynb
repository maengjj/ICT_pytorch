```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# MNIST 데이터셋 로드
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 데이터 전처리
x_train, x_test = x_train / 255.0, x_test / 255.0

# 모델 구성
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 훈련
model.fit(x_train, y_train, epochs=15)

# 모델 평가
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'\n테스트 정확도: {test_acc}')

# 테스트 이미지 선택 및 예측
num_images = 5
test_images = x_test[:num_images]
predictions = model.predict(test_images)

# 이미지와 예측 결과 시각화
fig, axes = plt.subplots(1, num_images, figsize=(20, 4))
for i, (image, prediction) in enumerate(zip(test_images, predictions)):
    axes[i].imshow(image, cmap='gray')
    predicted_label = np.argmax(prediction)
    true_label = y_test[i]
    axes[i].set_title(f'예측: {predicted_label}\n실제: {true_label}')
    axes[i].axis('off')

plt.tight_layout()
plt.show()

# 사용자 입력으로 특정 이미지 확인
while True:
    try:
        index = int(input("\n확인하고 싶은 테스트 이미지의 인덱스를 입력하세요 (0-9999, -1로 종료): "))
        if index == -1:
            break
        if 0 <= index < 10000:
            image = x_test[index]
            prediction = model.predict(image.reshape(1, 28, 28))[0]
            predicted_label = np.argmax(prediction)
            true_label = y_test[index]

            plt.figure(figsize=(6, 3))
            plt.subplot(1, 2, 1)
            plt.imshow(image, cmap='gray')
            plt.title(f'실제 숫자: {true_label}')
            plt.axis('off')

            plt.subplot(1, 2, 2)
            plt.bar(range(10), prediction)
            plt.title(f'예측 숫자: {predicted_label}')
            plt.xticks(range(10))
            plt.xlabel('숫자')
            plt.ylabel('확률')

            plt.tight_layout()
            plt.show()
        else:
            print("0에서 9999 사이의 숫자를 입력해주세요.")
    except ValueError:
        print("올바른 숫자를 입력해주세요.")
```
